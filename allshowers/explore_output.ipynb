```python
import showerdata


shower = showerdata.load("my_showers.h5")
```

```python
shower
```

```python
points=shower.points
energies=shower.energies
directions=shower.directions
pdg=shower.pdg
```

```python
print(points.shape)
print(energies.shape)
print(directions.shape)
print(pdg.shape)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
for p in points:
    plt.hist(p[:, 0], bins=10, alpha=0.5, label="x")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.title("Distribution of x-coordinates")
plt.show()


plt.figure(figsize=(10, 6))
for p in points:
    plt.hist(p[:, 1], bins=10, alpha=0.5, label="y")
plt.xlabel("y")
plt.ylabel("Frequency")
plt.title("Distribution of y-coordinates")
plt.show()


plt.figure(figsize=(10, 6))
for p in points:
    plt.hist(p[:, 2], bins=10, alpha=0.5, label="z")
plt.xlabel("z")
plt.ylabel("Frequency")
plt.title("Distribution of z-coordinates")
plt.show()



plt.figure(figsize=(10, 6))
for p in points:
    plt.hist(p[:, 3], bins=10, alpha=0.5, label="x")
plt.xlabel("e")
plt.ylabel("Frequency")
plt.title("Distribution of energies")
plt.show()



plt.figure(figsize=(10, 6))
for p in points:
    plt.hist(p[:, 4], bins=10, alpha=0.5, label="x")
plt.xlabel("t")
plt.ylabel("Frequency")
plt.title("Distribution of times")
plt.show()
```
